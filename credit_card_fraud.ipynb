```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.decomposition import TruncatedSVD
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.neighbors import LocalOutlierFactor
from sklearn.cluster import DBSCAN

%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
```

**Loading dataset**

```python
import kaggle
kaggle.api.authenticate() 
kaggle.api.dataset_download_files('mlg-ulb/creditcardfraud', path='data', unzip=True)
```

```python
data = pd.read_csv('data/creditcard.csv')
data.tail()
```

# EDA

```python
data.info()
```

```python
data.describe().round(2)
```

```python
print(f'{round(data['Class'].value_counts()[1] / len(data) * 100, 2)}% or {len(data.loc[data['Class'] == 1])} transactions are fraudulent.')
```

There are 492 fraudulent transactions in total, which is 0.17% of the total number of transactions.

## Distribution of data

```python
colors = ['#66c968', '#C96866']
```

```python
dist = data.drop('Class', axis=1)

plt.figure(figsize=(15, 15))
for i, col in enumerate(dist.columns):
    plt.subplot(6, 5, i+1).set_title(col)
    sns.boxplot(data=dist,
                x=dist.columns[i],
                hue=data['Class'],
                palette=colors)

plt.tight_layout()
plt.show()
```

```python
X = data.drop(['Class'], axis=1)
y = data['Class']

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

tsne = TSNE(n_components=2, random_state=RANDOM_SEED)
X_tsne = tsne.fit_transform(X_scaled)

svd = TruncatedSVD(n_components=2, random_state=RANDOM_SEED)
X_svd = svd.fit_transform(X_scaled)
```

```python
def decomp_plots(X_tsne, X_svd, hue=data['Class']):
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=hue, palette=colors, s=10)
    plt.title('t-SNE')

    plt.subplot(1, 2, 2)
    sns.scatterplot(x=X_svd[:, 0], y=X_svd[:, 1], hue=hue, palette=colors, s=10)
    plt.title('TruncatedSVD')

    plt.tight_layout()
    plt.show()

decomp_plots(X_tsne, X_svd)
```

# Models

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.25, random_state=RANDOM_SEED)

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Isolation Forest

```python
forest = IsolationForest(contamination=0.017, n_jobs=-1, random_state=RANDOM_SEED)
forest.fit(X_train)

y_forest = forest.predict(X_test)

y_forest = np.where(y_forest == 1, 0, y_forest)
y_forest = np.where(y_forest == -1, 1, y_forest)
```

```python
print(f'Anomalies are predicted: {sum(y_forest)}')
print(f'The proportion of predicted anomalies from the total number of transactions: {round(sum(y_forest)/len(data), 6) * 100}%')
print(f'Percentage of anomalies from the total number of transactions in the test sample: {round(sum(y_test)/len(X_test)* 100, 4)}%')
```

```python
print(classification_report(y_test, y_forest))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_forest)
plt.show()
```

```python
def compare_plot(y_pred):
    X_test_svd = svd.fit_transform(X_test)

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    sns.scatterplot(x=X_test_svd[:, 0], y=X_test_svd[:, 1], hue=y_pred, palette=colors, s=10)
    plt.title('Predicted data with TruncatedSVD')

    plt.subplot(1, 2, 2)
    sns.scatterplot(x=X_svd[:, 0], y=X_svd[:, 1], hue=data['Class'], palette=colors, s=10)
    plt.title('Original data with TruncatedSVD')

    plt.tight_layout()
    plt.show()
```

```python
compare_plot(y_forest)
```

There is a significant number of false-fraudulent transactions.

## One-Class SVM

```python
svm = OneClassSVM(nu=0.017)
svm.fit(X_train)

y_svm = svm.predict(X_test)

y_svm = np.where(y_svm == 1, 0, y_svm)
y_svm = np.where(y_svm == -1, 1, y_svm)
```

```python
print(f'Anomalies are predicted: {sum(y_svm)}')
print(f'The proportion of predicted anomalies from the total number of transactions: {round(sum(y_svm)/len(data) * 100, 4)}%')
print(f'Percentage of anomalies from the total number of transactions in the test sample: {round(sum(y_test)/len(X_test) * 100, 4)}%')
```

```python
print(classification_report(y_test, y_svm))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_svm)
plt.show()
```

```python
compare_plot(y_svm)
```

## Local Outlier Factor

```python
lof = LocalOutlierFactor(contamination=0.017, n_jobs=-1)

y_lof = lof.fit_predict(X_test)

# replacing values in the resulting array
y_lof = np.where(y_lof == 1, 0, y_lof)
y_lof = np.where(y_lof == -1, 1, y_lof)
```

```python
print(f'Anomalies are predicted: {sum(y_lof)}')
print(f'The proportion of predicted anomalies from the total number of transactions: {round(sum(y_lof)/len(data) * 100, 4)}%')
print(f'Percentage of anomalies from the total number of transactions in the test sample: {round(sum(y_test)/len(X_test) * 100, 4)}%')
```

```python
print(classification_report(y_test, y_lof))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_lof)
plt.show()
```

```python
compare_plot(y_lof)
```

In this case, the result is the opposite (but not in terms of its accuracy) - the model does not recognize many fraudulent transactions.

## DBSCAN

```python
outlier_percentage = 1.0
eps = 0.05
while outlier_percentage > 0.17:
    dbscan = DBSCAN(eps=eps).fit(X_test)
    labels = dbscan.labels_
    labels = np.array([1 if label == -1 else 0 for label in labels])
    outlier_percentage = np.sum(labels == 1) / len(labels)
    eps += 0.05

dbscan = DBSCAN(eps)
dbscan.fit(X_test)
y_dbscan = np.array([1 if label == -1 else 0 for label in dbscan.labels_])
```

```python
print(f'Anomalies are predicted: {sum(y_dbscan)}')
print(f'The proportion of predicted anomalies from the total number of transactions: {round(sum(y_dbscan)/len(data) * 100, 4)}%')
print(f'Percentage of anomalies from the total number of transactions in the test sample: {round(sum(y_test)/len(X_test) * 100, 4)}%')
```

```python
print(classification_report(y_test, y_dbscan))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_dbscan)
plt.show()
```

```python
compare_plot(y_dbscan)
```

This method works best of all.
